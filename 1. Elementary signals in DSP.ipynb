```python
# import necessary library
import matplotlib.pyplot as plt
import numpy as np
```

**Impulse / Unit Impulse**  
The unit impulse is denoted as $\delta$(n) and is defined as  
\begin{equation*}
\delta(n)   = {1,\;\;for\;\;n=0  \choose 0,\;\;for\;\:n\neq0}
\end{equation*}

```python
def plotUnitImpulse():
    sample = np.arange(-10,11,1)
    response = np.zeros(len(sample))
    response[10]=1
    plt.figure(figsize=(10, 6))
    plt.stem(sample,response)
    plt.xlabel('Sample')
    plt.ylabel('Amplitude')
    plt.title('Unit Impulse')
    plt.yticks([0,0.5,1])
    plt.xticks(sample)
    plt.grid('on')
    plt.show()
plotUnitImpulse()
```

**Ramp Signal**  
The unit ramp signal is denoted as $u_r$(n) and is defined as  
\begin{equation*}
u_r(n)   = {n,\;\;for\;\;n\ge0  \choose 0,\;\;for\;\:n\lt0}
\end{equation*}

```python
def plotRampSignal():
    sample = np.arange(-10,11,1)
    response = np.zeros(21)
    response[10]=1
    for i in range(10,21):
        response[i]=i-10
    plt.figure(figsize=(10, 6))
    plt.stem(sample,response)
    plt.xlabel('Sample')
    plt.xticks(sample)
    plt.ylabel('Amplitude')
    plt.title('Unit Ramp Signal')
    plt.grid('on')
    plt.show()
plotRampSignal()
```

**Exponential Signal**  
The exponential signal is denoted x(n) and is defined as  
\begin{equation*}
x(n)=a^n\;\;\;\;for\;all\;n
\end{equation*}

```python
def plotExponentialSignal():
    sample = np.arange(-10,11,1)
    a1 = 0.9    # 0 < a < 1
    a2 = 1.1    # a > 1
    a3 = -0.9   # -1 < a < 0
    a4 = -1.1   # a < -1

    x1 = a1**sample
    x2 = a2**sample
    x3 = a3**sample
    x4 = a4**sample
    # Plot  0 < a < 1
    plt.figure(figsize=(20, 16))
    plt.subplot(2,2,1)
    plt.stem(sample,x1)
    plt.xlabel('Sample')
    plt.xticks(sample)
    plt.ylabel('Amplitude')
    plt.title('Exponentian Signal (0 < a < 1)')
    # Plot  a > 1
    plt.subplot(2,2,2)
    plt.stem(sample,x2)
    plt.xlabel('Sample')
    plt.xticks(sample)
    plt.ylabel('Amplitude')
    plt.title('Exponentian Signal ( a > 1 )')
    # Plot  -1 < a < 0
    plt.subplot(2,2,3)
    plt.stem(sample,x3)
    plt.xlabel('Sample')
    plt.xticks(sample)
    plt.ylabel('Amplitude')
    plt.title('Exponentian Signal ( -1 < a < 0 )')
    # Plot  a < -1
    plt.subplot(2,2,4)
    plt.stem(sample,x4)
    plt.xlabel('Sample')
    plt.xticks(sample)
    plt.ylabel('Amplitude')
    plt.title('Exponentian Signal ( a < -1 )')
    plt.grid('on')
    plt.show()
plotExponentialSignal()
```

**Sinusoid Signals**  
Sinusoidal Wave Signal is a special type of signal. It is given by the function
  
\begin{equation*}
f(t)=Asin(2\Pi ft+\theta)
\end{equation*}

```python
def plotSinusoidSignal():
    A=5
    t = np.arange(-10,11,.001)
    theta = 0 
    f = 100
    response  = A*np.sin(2*np.pi*t+theta)
    plt.figure(figsize=(10, 8))
    plt.plot(t,response)
    plt.xlabel('time')
    plt.ylabel('Amplitude')
    plt.title('Sinusoid Signal')
    plt.grid('on')
    plt.show()
plotSinusoidSignal()
```
