```python
# import necessary library
import matplotlib.pyplot as plt
import numpy as np
```

**Convolution**  
Convolution is a mathematical operation used to express the relation between input and output of an LTI system. 
*Eqn. 2.3.17*
  
### Steps: 
1. Folding. Fold h(k) about k = 0 to obtain h(-k). 
2. Shifting. Shift h( -k) by no to the right (left) if no is positive (negative), to obtain h($n_o$ - k). 
3. Multiplication. Multiply x(k) by h($n_o$ - k) to obtain the product sequence  $v_no$(k) == x(k)h($n_o$ - k). 
4. Summation. Sum all the values of the product sequence $v_no$(k) to obtain the value of the output at time n = $n_o$. 



### Input and Impulse Response

```python
x=[1,2,3,1] #input signal origin 1
h=[1,2,1,-1] #impulse response origin 2
sample=np.arange(-10,11,1)
response=np.zeros(21)
for i in range(0,len(x),1):
    response[i+10]=x[i]
# Plot Input Response
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.title("Input Signal x(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,response)
# Plot impulse response
impulseResponse=np.zeros(21)
for i in range(0,len(h),1):
    impulseResponse[i+9]=h[i]

plt.figure(figsize=(20,10))
plt.subplot(2,2,3)
plt.title("Impulse Response h(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,impulseResponse)
plt.show()
```

**Folding the impulse response h(n) -> h(-n)**

```python
#Folding the impulse response
foldedResponse = np.flipud(impulseResponse)
plt.figure(figsize=(10,5))
plt.title("Folded Impulse Response h(-n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,foldedResponse)
plt.show()
```

**Shift the folded impulse , multiply with input ,store the summation**

```python
mul=[]
shiftedResponse=[]
ans=[]
for i in range(-1,6,1):
    # Shifiting
    shiftedResponse=np.roll(foldedResponse,i)
    #multiplication
    mul=np.multiply(shiftedResponse,response)
    #Summation
    ans.append(np.sum(mul))


print("Convolution of Two signal x(n) and h(n) :",ans)
print("origin : 2")
```

**Ploting Convolved signal**

```python
convolveResponse=np.zeros(21)
for i in range(0,len(ans),1):
    convolveResponse[9+i]=ans[i]

plt.figure(figsize=(20,10))
plt.title("Linear Convolution")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,convolveResponse)
plt.show()
```
