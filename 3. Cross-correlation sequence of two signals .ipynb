```python
# import necessary library
import matplotlib.pyplot as plt
import numpy as np
```

**Cross-correlation**  
The crosscorrelation of x(n) and y(n) is a sequence $r_{xy}(l)$, which is defined as  


```python
x=[2,-1,3,7,1,2,-3] # origin 5
y=[1,-1,2,-2,4,1,-2,5] # origin 5
sample=np.arange(-10,11,1)
rx=np.zeros(21)
for i in range(0,len(x),1):
    rx[i+6]=x[i]
# Plot X Response
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.title("Input Signal x(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,rx)
# Plot Y response
ry=np.zeros(21)
for i in range(0,len(y),1):
    ry[i+6]=y[i]
plt.figure(figsize=(20,10))
plt.subplot(2,2,3)
plt.title("Input Signal y(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,ry)
plt.show()
```

```python
mul=[]
shiftedResponse=[]
ans=[]
for i in range(-7,7,1):
    # Shifiting
    shiftedResponse=np.roll(ry,i)
    #multiplication
    mul=np.multiply(shiftedResponse,rx)
    #Summation
    ans.append(np.sum(mul))
print("Cross-Correlation : ",ans)
```

```python
crossCorrelation=np.zeros(21)
for i in range(0,len(ans),1):
    crossCorrelation[3+i]=ans[i]

plt.figure(figsize=(20,10))
plt.title("Cross Correlation")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,crossCorrelation)
plt.show()
```
