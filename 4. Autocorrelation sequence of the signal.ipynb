```python
# import necessary library
import matplotlib.pyplot as plt
import numpy as np
```

**Autocorrelation**  
The autocorrelation of x(n) defined as  


```python
x=[2,-1,3,7,1,2,-3] # origin 5
y=[2,-1,3,7,1,2,-3] # origin 5
sample=np.arange(-10,11,1)
rx=np.zeros(21)
for i in range(0,len(x),1):
    rx[i+6]=x[i]
# Plot X Response
ry=rx
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.title("Input Signal x(n)")
plt.xlabel("Sample")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,rx)
plt.show()
```

```python
mul=[]
shiftedResponse=[]
ans=[]
for i in range(-7,8,1):
    # Shifiting
    shiftedResponse=np.roll(ry,i)
    #multiplication
    mul=np.multiply(shiftedResponse,rx)
    #Summation
    ans.append(np.sum(mul))
print("Auto-Correlation : ",ans)
```

```python
crossCorrelation=np.zeros(21)
for i in range(0,len(ans),1):
    crossCorrelation[3+i]=ans[i]

plt.figure(figsize=(20,10))
plt.title("Auto Correlation")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.xticks(sample)
plt.stem(sample,crossCorrelation)
plt.show()
```
